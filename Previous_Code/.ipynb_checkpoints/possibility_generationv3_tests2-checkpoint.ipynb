```python
import torch
import transformers
from transformers import BertForSequenceClassification, XLMRobertaForSequenceClassification, AdamW, BertConfig, BertTokenizer, XLMRobertaTokenizer
from torch.utils.data import TensorDataset, DataLoader
import numpy as np
from google.colab import drive
import re
from bs4 import BeautifulSoup
```

```python
'''
Construct object in BERT sentiment classifier class
'''
sentiment_model = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased", # Use the 12-layer BERT model, with an uncased vocab.
    num_labels = 2, # The number of output labels--2 for binary classification.
                    # You can increase this for multi-class tasks.   
    output_attentions = False, # Whether the model returns attention weights.
    output_hidden_states = False # Whether the model returns all hidden states.
)

sentiment_model.cuda() # Puts model on GPU
```

```python
'''
Load pretrained sentiment model parameters
'''
sentiment_model.load_state_dict(torch.load("/content/drive/My Drive/English_sentiment_model.pt", map_location=torch.device('cpu')), strict=False)
```

```python
sentiment_model.eval() # Put model in inference (as opposed to training) mode
```

```python
'''
Load tokenizer for sentiment model
'''
sentiment_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
```

```python
# gets the text into the format we want
def clean_tweet(tweet):
  tweet = BeautifulSoup(tweet, "lxml").get_text() # turns xml-formatted text into regular text
  tweet = re.sub(r"@[A-Za-z0-9]+", " ", tweet) # gets rid of all user references in tweets (i.e. "@username")
  tweet = re.sub(r"https?://[A-Za-z0-9./]+", " ", tweet) # gets rid of URLs
  tweet = re.sub(r"[^A-Za-z.!?áéíóúüñ¿ÁÉÍÓÚÜÑ']", " ", tweet) # gets rid of any non-standard characters in the tweets
  tweet = re.sub(r" +", " ", tweet) # replaces all excess whitespace with a single space

  return tweet # gives us our cleaned tweet
```

```python
def predict_sentiment(tweet):

    '''
    Predict the sentiment of an input text
    '''
    
    tweet = clean_tweet(tweet)
    tweet_input_id = []
    tweet_attention_mask = []

    tweet_dict = sentiment_tokenizer.encode_plus(
                            tweet,                      # Sentence to encode.
                            add_special_tokens = True, # Add '[CLS]' and '[SEP]'
                            max_length = 64,           # Pad & truncate all sentences.
                            truncation=True,           # Explicitly enable truncation
                            padding = 'max_length',
                            return_attention_mask = True,   # Construct attn. masks.
                            return_tensors = 'pt',     # Return pytorch tensors.
                       )

    # Add the encoded sentence to the list.    
    tweet_input_id.append(tweet_dict['input_ids'])

    # And its attention mask (simply differentiates padding from non-padding).
    tweet_attention_mask.append(tweet_dict['attention_mask'])

    # Convert the lists into tensors.
    tweet_input_id = torch.cat(tweet_input_id, dim=0)
    tweet_attention_mask = torch.cat(tweet_attention_mask, dim=0)

    tweet_data = TensorDataset(tweet_input_id, tweet_attention_mask)
    
    tweet_dataloader = DataLoader(tweet_data)
    
    for data in tweet_dataloader:
        tweet_input_id = data[0].to(device=torch.device('cuda'))
        tweet_attention_mask = data[1].to(device=torch.device('cuda'))
    
    tweet_logits = sentiment_model(tweet_input_id, token_type_ids=None, attention_mask=tweet_attention_mask)
    
    tweet_logits = tweet_logits[0].detach().cpu().numpy()

    tweet_logits = torch.Tensor(tweet_logits)

    softmax = torch.nn.Softmax(dim=1)
    prob_dist = softmax(tweet_logits)

    sentiment_pred = prob_dist.tolist()

    sentiment_pred = sentiment_pred[0][1]

    return sentiment_pred
```

```python
import pandas as pd 
import numpy as np 
import re

df = pd.read_csv('https://raw.githubusercontent.com/gokulsrin/PossibilityGeneration/main/TimeData/generationLD%20(1).csv')
```

```python
from transformers import pipeline
predict_sentiment = pipeline("sentiment-analysis")
```

```python
g = df.groupby('ResponseId')
#{ind: {scenario: {answers: [answers], times:[times]} }
data = {} 
#by person
for id, group in g:
 g2 = group.groupby('scenario')
 data[id] = {}
 #by scenario
 for scenario, group2 in g2:
   data.get(id)[scenario] = {'answers':[], 'times': []}
   g3 = group2.groupby('answerOrder')
   #it is sorted by order correctly 
   for answerOrder, group3 in g3:
      row = group3.iloc[0]
      data.get(id).get(scenario).get('answers').append(row['question'])
      data.get(id).get(scenario).get('times').append(row['time'])
      continue 
```

```python
#fixing a scenario to do the sentiment of generation vs. time plot 

time = []
sentiment = []

for id in data:
  for s in data.get(id):
    # if s == 'scenario1' or s == 'scenario2' or s == 'scenario3':
      for i in range(len(data.get(id).get(s).get('answers'))):
        sentiment.append(predict_sentiment(data.get(id).get(s).get('answers')[i])[0].get('score'))
        time.append(data.get(id).get(s).get('times')[i])
```

```python
import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats

# i = 0 
# while i < len(time):
#   if time[i] > 300:
#     time.pop(i)
#     sentiment.pop(i)
#     i -= 1
#   i +=1
x = np.array(time)
y = np.array(sentiment)

plt.scatter(x, y)
plt.xlabel("Generation Time")
plt.ylabel("Generation Sentiment")
plt.title("All Scenarios")
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
predict_y = intercept + slope * x

plt.plot(x,y,'ro')
plt.plot(x, predict_y, 'k-')
plt.autoscale(True)
plt.grid(True)
plt.show()

print("r-value:%f" % r_value)
print("p-value:%f" % p_value)
```

```python
#another test - seeing if the difference in sentiment between the 1st & 8th possibility is related to the amount of time elapsed in between sum(time[2:8])

#fixing a scenario to do the sentiment of generation vs. time plot 

time_difference = []
sentiment_difference = []

for id in data:
  for s in data.get(id):
    # if sum(data.get(id).get(s).get('times')[1:]) < 600:
      sentiment_difference.append(abs(predict_sentiment(data.get(id).get(s).get('answers')[0])[0].get('score') - predict_sentiment(data.get(id).get(s).get('answers')[-1])[0].get('score')))
      time_difference.append(sum(data.get(id).get(s).get('times')[1:]))
```

```python
x = np.array(time_difference)
y = np.array(sentiment_difference)

plt.scatter(x, y)
plt.xlabel("Time Difference")
plt.ylabel("Sentiment Difference")
plt.title("All Scenarios")
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
predict_y = intercept + slope * x

plt.plot(x,y,'ro')
plt.plot(x, predict_y, 'k-')
plt.autoscale(True)
plt.grid(True)
plt.show()

print("r-value:%f" % r_value)
print("p-value:%f" % p_value)
```

```python
#find average sentiment within vignette, then total time it took to get that 
total_time = []
avg_sentiment = []

for id in data:
  for s in data.get(id):
      p = 0 
      for ans in data.get(id).get(s).get('answers'):
        p += predict_sentiment(ans)[0].get('score')
      p/= len(data.get(id).get(s).get('answers'))
      avg_sentiment.append(p)
      total_time.append(sum(data.get(id).get(s).get('times')))
```
