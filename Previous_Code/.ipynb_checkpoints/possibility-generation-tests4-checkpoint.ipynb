```python
'''
Load tokenizer for sentiment model
'''
sentiment_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
```

```python
# gets the text into the format we want
def clean_tweet(tweet):
  tweet = BeautifulSoup(tweet, "lxml").get_text() # turns xml-formatted text into regular text
  tweet = re.sub(r"@[A-Za-z0-9]+", " ", tweet) # gets rid of all user references in tweets (i.e. "@username")
  tweet = re.sub(r"https?://[A-Za-z0-9./]+", " ", tweet) # gets rid of URLs
  tweet = re.sub(r"[^A-Za-z.!?áéíóúüñ¿ÁÉÍÓÚÜÑ']", " ", tweet) # gets rid of any non-standard characters in the tweets
  tweet = re.sub(r" +", " ", tweet) # replaces all excess whitespace with a single space

  return tweet # gives us our cleaned tweet
```

```python
def predict_sentiment(tweet):

    '''
    Predict the sentiment of an input text
    '''
    
    tweet = clean_tweet(tweet)
    tweet_input_id = []
    tweet_attention_mask = []

    tweet_dict = sentiment_tokenizer.encode_plus(
                            tweet,                      # Sentence to encode.
                            add_special_tokens = True, # Add '[CLS]' and '[SEP]'
                            max_length = 64,           # Pad & truncate all sentences.
                            truncation=True,           # Explicitly enable truncation
                            padding = 'max_length',
                            return_attention_mask = True,   # Construct attn. masks.
                            return_tensors = 'pt',     # Return pytorch tensors.
                       )

    # Add the encoded sentence to the list.    
    tweet_input_id.append(tweet_dict['input_ids'])

    # And its attention mask (simply differentiates padding from non-padding).
    tweet_attention_mask.append(tweet_dict['attention_mask'])

    # Convert the lists into tensors.
    tweet_input_id = torch.cat(tweet_input_id, dim=0)
    tweet_attention_mask = torch.cat(tweet_attention_mask, dim=0)

    tweet_data = TensorDataset(tweet_input_id, tweet_attention_mask)
    
    tweet_dataloader = DataLoader(tweet_data)
    
    for data in tweet_dataloader:
        tweet_input_id = data[0].to(device=torch.device('cuda'))
        tweet_attention_mask = data[1].to(device=torch.device('cuda'))
    
    tweet_logits = sentiment_model(tweet_input_id, token_type_ids=None, attention_mask=tweet_attention_mask)
    
    tweet_logits = tweet_logits[0].detach().cpu().numpy()

    tweet_logits = torch.Tensor(tweet_logits)

    softmax = torch.nn.Softmax(dim=1)
    prob_dist = softmax(tweet_logits)

    sentiment_pred = prob_dist.tolist()

    sentiment_pred = sentiment_pred[0][1]

    return sentiment_pred
```

```python
embeddings = []
res = []
transition_map = {}
for scenario, group in generations.groupby("vignette_name"):
  #Mary is good
  # if scenario == "Mary":
    for id, group1 in group.groupby("id"):
      responses = group1['response'].values.tolist()
      # for pnum, group2 in group1.groupby("generation_number"):
      for i in range(len(responses)):
        response = responses[i]
        if not pd.isna(response):
          res.append(response)
          embeddings.append(model.encode(response))
          if i < len(responses)-1 and not pd.isna(responses[i+1]):
            transition_map[response] = responses[i+1]
  # break
```

```python
# t{cluster1} = {cluster2: probability}
# transition probability from cluster 1 to cluster 2
transitions = {}

for i in range(len(df.values)):
  r1 = res[i]
  if r1 in transition_map:
    r2 = transition_map[r1]
  if cluster_dict[r1] not in transitions:
    transitions[cluster_dict[r1]] = {}
  if cluster_dict[r2] not in transitions[cluster_dict[r1]]:
    transitions[cluster_dict[r1]][cluster_dict[r2]] = 1
  else:
    transitions[cluster_dict[r1]][cluster_dict[r2]] += 1
#normalize transitions 
for c1 in transitions:
  s = 0 
  for c2 in transitions[c1]:
    s += transitions[c1][c2]
  for c2 in transitions[c1]:
    transitions[c1][c2] /= s
    transitions[c1][c2] = round(transitions[c1][c2], 4)
```

```python
# for visualizing cluster transitions as a graph 
import networkx as nx
G = nx.Graph()
#vertices 
for c in transitions:
  G.add_node(c)
#edges 
for c1 in transitions:
  for c2 in transitions[c1]:
    G.add_edge(c1, c2, weight = transitions[c1][c2])
```

```python
vignette_order = []
for vign in id_trajectories[inds[0]]:
  vignette_order.append(vign)
vignette_order = vignette_order[:-1]
last_vign = "Mary"
```

```python
#ask to what extent similarity in clusters trajectories between two particiapnts for n-1 vignettes (avg diff) can predict similarity in nth vignette
# could look at median value
# the plot changes drastically when the avg for all 10 is used to predict 1 
x = []
y = []
for i in range(len(inds)):
  for j in range(i+1,len(inds)):
    #ensure they have the same # of vigns 
    if len(id_trajectories[inds[i]]) == len(id_trajectories[inds[j]]) == 10:
      avg_sim = []
      #for the first 9 vignettes, find the average cosine sim between trajectory vectors
      for vign in vignette_order:
        #cosine sim
        a = id_trajectories[inds[i]][vign]
        b = id_trajectories[inds[j]][vign]
        cos_sim = dot(a, b)/(norm(a)*norm(b))
        #ensure this is not NAN
        if not pd.isna(cos_sim):
          avg_sim.append(cos_sim)
      # if the avg_sim we calculated was valid
      if len(avg_sim) > 0:
        avg_sim = sum(avg_sim)/len(avg_sim)
        #find the cos sim of the last vign
        a = id_trajectories[inds[i]][last_vign]
        b = id_trajectories[inds[j]][last_vign]
        cos_sim = dot(a, b)/(norm(a)*norm(b))
        #ensure it is not NAN
        if not pd.isna(cos_sim) and cos_sim > 0 and cos_sim < 1 and avg_sim > 0:
          x.append(avg_sim)
          y.append(cos_sim)    
```

```python
#ask to what extent similarity in clusters trajectories between two particiapnts for n-1 vignettes (avg diff) can predict similarity in nth vignette
x = []
y = []
for i in range(len(inds)):
  for j in range(i+1,len(inds)):
    #ensure they have the same # of vigns 
    if len(id_trajectories[inds[i]]) == len(id_trajectories[inds[j]]) == 10:
      avg_sim = []
      #for the first 9 vignettes, find the average cosine sim between trajectory vectors - currently all 10 
      for vign in vignette_order:
        #cosine sim
        a = id_trajectories[inds[i]][vign][0]
        b = id_trajectories[inds[j]][vign][0]
        cos_sim = dot(a, b)/(norm(a)*norm(b))
        #ensure this is not NAN
        if not pd.isna(cos_sim):
          avg_sim.append(cos_sim)
      # if the avg_sim we calculated was valid
      if len(avg_sim) > 0:
        avg_sim = sum(avg_sim)/len(avg_sim)
        #find the sentiment difference 
        a = id_trajectories[inds[i]][vign][1]
        b = id_trajectories[inds[j]][vign][1]
        a = sum(a)/len(a)
        b = sum(b)/len(b)
        x.append(avg_sim)
        y.append(a-b)
```

```python
# if we remove y = 0 data, what happens

# a = np.random.rand(10)
# b = a * 2 + a ** 5

# print("The Rho is {}".format(np.corrcoef(a, b)[0][1]))
g = sns.jointplot(x=x, y=y, kind='reg', color='royalblue')
# ax.annotate(stats.pearsonr)
r, p = stats.pearsonr(x, y)
g.ax_joint.annotate(f'$\\rho = {r:.3f}, p = {p:.3f}$',
                    xy=(0.1, 0.9), xycoords='axes fraction',
                    ha='left', va='center',
                    bbox={'boxstyle': 'round', 'fc': 'powderblue', 'ec': 'navy'})
g.ax_joint.scatter(x, y)
g.set_axis_labels(xlabel='Average Cosine Similarity', ylabel='Sentiment Difference', size=12)
plt.tight_layout()
plt.show()
```

```python
concreteness = pd.read_csv("https://raw.githubusercontent.com/gokulsrin/PossibilityGeneration/main/Misc/Concreteness_ratings_Brysbaert_et_al_BRM.csv")
```

```python
#create concreteness dict
concreteness_dict = {}
for word, group in concreteness.groupby("Word"):
  rating = group["Conc.M"].values[0]
  concreteness_dict[word] = rating
```

```python
#figure out if later generations are more or less concrete 
import string 

conc = []
nums = []
cdic = {}
for id, group in generations.groupby("id"):
  for vign, group2 in group.groupby("vignette_name"):
    for num, group3 in group2.groupby("generation_number"):
      response = group3["response"].values[0]
      if not pd.isna(response):
        response = response.translate(str.maketrans('', '', string.punctuation))
        score = []
        l = len(response.split(" "))
        for word in response.split(" "):
          if word in concreteness_dict:
            score.append(concreteness_dict[word])
        if len(score) > 0:
          score = sum(score)/len(score)
          # score = (sum(score)/len(score))/l
          conc.append(score)
          nums.append(num)
          if num not in cdic:
            cdic[num] = []
          cdic[num].append(score)
```

```python
#find averages in concreteness differences 
for key in cdic:
  cdic[key] = sum(cdic[key])/len(cdic[key])

print(cdic)
```
