```python
'''
Load tokenizer for sentiment model
'''
sentiment_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
```

```python
# gets the text into the format we want
def clean_tweet(tweet):
  tweet = BeautifulSoup(tweet, "lxml").get_text() # turns xml-formatted text into regular text
  tweet = re.sub(r"@[A-Za-z0-9]+", " ", tweet) # gets rid of all user references in tweets (i.e. "@username")
  tweet = re.sub(r"https?://[A-Za-z0-9./]+", " ", tweet) # gets rid of URLs
  tweet = re.sub(r"[^A-Za-z.!?áéíóúüñ¿ÁÉÍÓÚÜÑ']", " ", tweet) # gets rid of any non-standard characters in the tweets
  tweet = re.sub(r" +", " ", tweet) # replaces all excess whitespace with a single space

  return tweet # gives us our cleaned tweet
```

```python
def predict_sentiment(tweet):

    '''
    Predict the sentiment of an input text
    '''
    
    tweet = clean_tweet(tweet)
    tweet_input_id = []
    tweet_attention_mask = []

    tweet_dict = sentiment_tokenizer.encode_plus(
                            tweet,                      # Sentence to encode.
                            add_special_tokens = True, # Add '[CLS]' and '[SEP]'
                            max_length = 64,           # Pad & truncate all sentences.
                            truncation=True,           # Explicitly enable truncation
                            padding = 'max_length',
                            return_attention_mask = True,   # Construct attn. masks.
                            return_tensors = 'pt',     # Return pytorch tensors.
                       )

    # Add the encoded sentence to the list.    
    tweet_input_id.append(tweet_dict['input_ids'])

    # And its attention mask (simply differentiates padding from non-padding).
    tweet_attention_mask.append(tweet_dict['attention_mask'])

    # Convert the lists into tensors.
    tweet_input_id = torch.cat(tweet_input_id, dim=0)
    tweet_attention_mask = torch.cat(tweet_attention_mask, dim=0)

    tweet_data = TensorDataset(tweet_input_id, tweet_attention_mask)
    
    tweet_dataloader = DataLoader(tweet_data)
    
    for data in tweet_dataloader:
        tweet_input_id = data[0].to(device=torch.device('cuda'))
        tweet_attention_mask = data[1].to(device=torch.device('cuda'))
    
    tweet_logits = sentiment_model(tweet_input_id, token_type_ids=None, attention_mask=tweet_attention_mask)
    
    tweet_logits = tweet_logits[0].detach().cpu().numpy()

    tweet_logits = torch.Tensor(tweet_logits)

    softmax = torch.nn.Softmax(dim=1)
    prob_dist = softmax(tweet_logits)

    sentiment_pred = prob_dist.tolist()

    sentiment_pred = sentiment_pred[0][1]

    return sentiment_pred
```

```python
import math 
def dist(r1, r2):
  #literally just sqrt(sum(i = 0 -> n, (r1[i] - r2[i])^2)), or the euclidian distance between these two embedding vectors
  d = [(a - b)**2 for a, b in zip(r1, r2)]
  d = sum(d)**.5
  return d
```

```python
#embedding dic - map all a participants embeddings to them
embed_dic = {}
```

```python
# finding the semantic distance/exploration for each individual
# this means I sum up the intravignette distance and then average across all 10 vignettes... 
sdist = {}
maxdist = {}
for id, group in generations.groupby('id'):
  maxdist[id] = 0
  distance = 0
  if id not in embed_dic:
    embed_dic[id] = {}
  for vign, group2 in group.groupby('vignette_name'):
    embeddings = []
    for response in group2['response'].tolist():
      #make sure response is nonnull
      if not pd.isna(response):
        embeddings.append(model.encode(response))
    embed_dic.get(id)[vign] = embeddings 
    for i in range(len(embeddings)):
      for j in range(i+1, len(embeddings)):
        d = dist(embeddings[i], embeddings[j])
        distance += d
        maxdist[id] = max(maxdist.get(id), d)
  distance /= 10
  sdist[id] = distance 
print(sdist)
```

```python
def average_embedding(embeddings):
  df = pd.DataFrame(embeddings)
  df = df.mean(axis=0)
  return df.values.tolist()
```

```python
# Center of mass semantic distance vs time to generate 
DCOM = []
time = []
for id, group in generations.groupby('id'):
  #for each vign
  for vign, group1 in group.groupby('vignette_name'):
    embeddings = []
    #for every response
    for row in zip(group1['response'].tolist(), group1['generation_time'].tolist()):
      response = row[0]
      gtime = row[1]
      #make sure response is nonnull
      if not pd.isna(response):
        if len(embeddings) > 0:
          current = model.encode(response)
          COM = average_embedding(embeddings)
          DCOM.append(dist(current, COM))
          time.append(gtime)
        embeddings.append(model.encode(response))
```

```python
x = np.array(DCOM)
y = np.array(time)
y = y/1000

#outlier removal
i = 0
while i < len(x):
  if y[i] > 200 or y[i] == 0 or x[i] == 0:
    y = np.delete(y, i)
    x = np.delete(x, i)
    i -= 1
  i += 1

plt.scatter(x, y)
plt.xlabel("Distance from Center of Mass")
plt.ylabel("Time (s)")

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
predict_y = intercept + slope * x

plt.plot(x,y,'ro')
plt.plot(x, predict_y, 'k-')
plt.autoscale(True)
plt.grid(True)
plt.show()

print("r-value:%f" % r_value)
print("p-value:%f" % p_value)
```

```python
def total_variation(e):
  variation = 0
  count = 0
  for i in range(len(e)):
    for j in range(i+1, len(e)):
      variation += dist(e[i], e[j])
      count += 1
      if count%10000==0:
        print("done")
  return variation
```

```python
#going to try to cluster all embeddings and see which sequential embeddings leave cluster vs stay
#response cluster = {response: cluster num}
rc = {}
responses = []

for id, group in generations.groupby("id"):
  for vign, group2 in group.groupby("vignette_name"):
    for response in group2["response"]:
      if not pd.isna(response):
        responses.append(response)
```

```python
#transform to embeddings
embeddings = []
for response in responses:
  embeddings.append(model.encode(response))
```

```python
from sklearn.cluster import KMeans
space = np.array(embeddings)
num_clusters = 8
kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(space)
labels = kmeans.labels_
```

```python
#map to response cluster
for i in range(len(responses)):
  rc[responses[i]] = labels[i]
```

```python
# partition will map {response: i}, where i is 1,0 depending on whether or not the response belongs to the same cluster as its predecessor
partition = {}
for id, group in generations.groupby("id"):
  for vign, group2 in group.groupby("vignette_name"):
    rs = list(zip(group2["response"], group2["generation_time"]))
    for i in range(1, len(rs)):
      # if the ith and i-1th response belong to the same cluster, indicate this with 1 
      if not pd.isna(rs[i][0]) and not pd.isna(rs[i-1][0]):
        if rc[rs[i][0]] == rc[rs[i-1][0]]:
          partition[rs[i]] = (1, rs[i][1])
        else:
          partition[rs[i]] = (0, rs[i][1])
```

```python
i = []
o = []

for key in partition:
  if partition[key][0] == 0:
    o.append(partition[key][1])
  else:
    i.append(partition[key][1])

iavg = sum(i)/len(i)
oavg = sum(o)/len(o)
```

```python
times = []
ks = []
for i in range(3, 15):
  #cluster
  space = np.array(embeddings)
  num_clusters = i
  kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(space)
  labels = kmeans.labels_
  #map to response cluster
  for i in range(len(responses)):
    rc[responses[i]] = labels[i]
  # partition will map {response: i}, where i is 1,0 depending on whether or not the response belongs to the same cluster as its predecessor
  partition = {}
  for id, group in generations.groupby("id"):
    for vign, group2 in group.groupby("vignette_name"):
      rs = list(zip(group2["response"], group2["generation_time"]))
      for i in range(1, len(rs)):
        # if the ith and i-1th response belong to the same cluster, indicate this with 1 
        if not pd.isna(rs[i][0]) and not pd.isna(rs[i-1][0]):
          if rc[rs[i][0]] == rc[rs[i-1][0]]:
            partition[rs[i]] = (1, rs[i][1])
          else:
            partition[rs[i]] = (0, rs[i][1])
  #find avg generation time
  i = []
  o = []
  for key in partition:
    if partition[key][0] == 0:
      o.append(partition[key][1])
    else:
      i.append(partition[key][1])
  iavg = sum(i)/len(i)
  oavg = sum(o)/len(o)
  # find diff between same and different
  times.append(iavg - oavg)
  ks.append(num_clusters)
```

```python
y = np.array(times)
x = np.array(ks)

plt.scatter(x, y)
plt.ylabel("Time Difference (ms)")
plt.xlabel("K (Cluster Granularity)")

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
predict_y = intercept + slope * x

plt.plot(x,y,'ro')
plt.plot(x, predict_y, 'k-')
plt.autoscale(True)
plt.grid(True)
plt.show()

print("r-value:%f" % r_value)
print("p-value:%f" % p_value)
```

```python
#figure out if later generations are more or less concrete 
import string 

conc = []
nums = []
cdic = {}
for id, group in generations.groupby("id"):
  for vign, group2 in group.groupby("vignette_name"):
    for num, group3 in group2.groupby("generation_number"):
      response = group3["response"].values[0]
      if not pd.isna(response):
        response = response.translate(str.maketrans('', '', string.punctuation))
        score = []
        l = len(response.split(" "))
        for word in response.split(" "):
          if word in concreteness_dict:
            score.append(concreteness_dict[word])
        if len(score) > 0:
          score = sum(score)/len(score)
          # score = (sum(score)/len(score))/l
          conc.append(score)
          nums.append(num)
          if num not in cdic:
            cdic[num] = []
          cdic[num].append(score)
```
